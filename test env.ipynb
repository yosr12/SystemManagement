```python
'''
This file is to perfom some method beforme implementing them into   the projects 
'''
import csv

def add_column_to_csv(csv_file, new_column_name, new_column_data):
    with open(csv_file, 'r') as file:
        reader = csv.DictReader(file)
        fieldnames = reader.fieldnames + [new_column_name]

        rows = []
        for row in reader:
            row[new_column_name] = new_column_data
            rows.append(row)

    with open(csv_file, 'w', newline='') as file:
        writer = csv.DictWriter(file, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(rows)

# Example usage:
csv_file = 'profiles_dataset.csv'
new_column_name = 'batteryLevel'
new_column_data = 10

add_column_to_csv(csv_file, new_column_name, new_column_data)
```

```python
def add_value_to_last_line(csv_file, column_name, new_value):
    rows = []
    with open(csv_file, 'r') as file:
        reader = csv.DictReader(file)
        fieldnames = reader.fieldnames

        for row in reader:
            rows.append(row)

    if column_name not in fieldnames:
        fieldnames.append(column_name)

    rows[-1][column_name] = new_value

    with open(csv_file, 'w', newline='') as file:
        writer = csv.DictWriter(file, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(rows)

csv_file = 'profiles_dataset.csv'
column_name = 'batteryLevel'
new_value = 1000

add_value_to_last_line(csv_file, column_name, new_value)
```
